```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as pl

pil_im = Image.open('../data/empire.jpg').convert('L')
                    
pl.figure()
pl.imshow(pil_im)
pl.gray()
pl.axis('off')
pl.show()
```

```python
help(Image.open)
```

```python
import os
filelist = os.listdir('../data')
for infile in filelist:
    outfile = os.path.splitext(infile)[0] + ".jpg"
    print(outfile)
    if infile != outfile:
        try:
            Image.open(infile).save(outfile)
        except IOError:
            print("cannot convert", infile)
```

```python
box = (150,150,500,500)
region = pil_im.crop(box)
pl.figure()
pl.imshow(region)
pl.gray()
pl.axis('on')
pl.show()
```

```python
print(region)
```

```python
def showimage(src):
    pl.figure()
    pl.imshow(src)
    pl.axis('on')
    pl.show()
```

```python
showimage(region)
```

```python
region = region.transpose(Image.ROTATE_90)
showimage(region)
```

```python
pil_im.paste(region,box)
showimage(pil_im)
```

```python
out = pil_im.resize((256,256))
showimage(out)
```

```python
out = out.rotate(60)
showimage(out)
```
