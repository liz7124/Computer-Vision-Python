```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('images/images03.jpg')
plt.rcParams["figure.figsize"] = (16,9)
```

```python
arr_kernel = [[1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[1,4,6,4,1],[4,16,24,16,4]]

kernel = np.array((arr_kernel),np.float32)/256
dst = cv2.filter2D(img,-1,kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(dst),plt.title('Result')
plt.show()
```
